```python
# update the "CISA Score" in df_renamed only if a new value is available in df_CISA_Source

# Create a mapping from the source file
cisa_mapping = df_CISA_Source.set_index("Project Name")["CISA Score"]
# Update only if a new CISA Score is available
df_renamed["CISA Score"] = df_renamed["Project Name"].map(cisa_mapping).combine_first(df_renamed["CISA Score"])
```

```python
#verification of the updated CISA Score

# Merge to compare the CISA scores before and after updating
df_check = df_renamed.merge(df_CISA_Source[["Project Name", "CISA Score"]],
                            on="Project Name", how="left", suffixes=("_renamed", "_source"))

# Identify rows where the updated CISA Score (renamed) doesn't match the source
df_mismatched = df_check[(df_check["CISA Score_source"].notna()) & 
                         (df_check["CISA Score_renamed"] != df_check["CISA Score_source"])]

# Display mismatches
print(f"Total mismatched rows: {len(df_mismatched)}")
print(df_mismatched)
```

```python
#First round, three buckets
#**Bucket 1**: `Priority_Grant_Funding`, Projects where County Score, CISA Score, and Mitigation Score are all equal to 1.
#**Bucket 2**: `Critical_Infrastructure_Mitigation`, Projects where County Score and Mitigation Score are equal to 1, but CISA Score is equal to 0.
#**Bucket 3**: `Other_Projects`, All other projects not meeting the conditions for Bucket 1 or Bucket 2

# Bucket 1: County Score, CISA Score, and Mitigation Score are all equal to 1
bucket1 = df_renamed[(df_renamed['County Score'] == 1) &
                     (df_renamed['CISA Score'] == 1) &
                     (df_renamed['Mitigation Score'] == 1)]

# Bucket 2: County Score and Mitigation Score are 1, CISA Score is 0
bucket2 = df_renamed[(df_renamed['County Score'] == 1) &
                     (df_renamed['Mitigation Score'] == 1) &
                     (df_renamed['CISA Score'] == 0)]
# Bucket 3: All other projects not in Bucket 1 or Bucket 2
bucket3 = df_renamed[~df_renamed.index.isin(bucket1.index) & 
                     ~df_renamed.index.isin(bucket2.index)]

# Output or save the buckets (optional)
# Saving the buckets as separate files if needed
bucket1.to_excel(r'C:\Users\bye\OneDrive - County of San Mateo\Desktop\0.CRISP\CRISP Data Collection\Project Submission\CRISP_Priority_Grant_Funding_v1_20250213.xlsx', index=False)
bucket2.to_excel(r'C:\Users\bye\OneDrive - County of San Mateo\Desktop\0.CRISP\CRISP Data Collection\Project Submission\CRISP_Critical_Infrastructure_Mitigation_v1_20250213.xlsx', index=False)
bucket3.to_excel(r'C:\Users\bye\OneDrive - County of San Mateo\Desktop\0.CRISP\CRISP Data Collection\Project Submission\CRISP_Other_Projects_v1_20250213.xlsx', index=False)

# Print the sizes of each bucket to verify
print(f"Bucket 1 size: {len(bucket1)}")
print(f"Bucket 2 size: {len(bucket2)}")
print(f"Bucket 3 size: {len(bucket3)}")
```
